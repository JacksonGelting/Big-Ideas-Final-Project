```python
# first lets import Pandas!
import pandas as pd
```

```python
# great! Now let's work.
# We have to import the CSV file that contains all of the information about housing. I set this as 'housingData'

housingData = pd.read_csv('housing.csv')

# let's see what the default data set looks like

housingData
```

```python
# woah! that's a lot of data. Lot's of columns and rows. A lot of these we don't want to deal with.
housingData.drop(['Sector #','Population (2010)','Total # Units (2000)','# Occupied Units (2010)','% Owner Occupied Units (2010)','% Renter Occupied Units (2010)','Est. Avg. Yrs. of Residence (2010)','% Living in Household (2010)','% Living in Group Quarters (2010)','% Units Built 00-09','% Units Built 90-99','% Units Built 80-89','% Units Built 60-79','% Units Built 40-59','% Units Built before 1939','Median Home Value (2000)','Median Home  Value (2010)','% Change Real Value 2000-2010','Median Sale Price (2010)','# Sales Counted (2010)','Foreclosures (2008)','Foreclosures (2010)','% of all Housing Units Foreclosed (2010)'],axis=1,inplace=True)

# NOW let's see what that looks like after dropping most of the columns
housingData
```

```python
# next, we have to sort the values by the percentage of homes occupied in the neighborhood.

clearedData = housingData.sort_values(by=['% Occupied Units (2010)'], ascending = [False]).reset_index(drop=True)

# let's see what that looks like

clearedData
```

```python
# what's that all the way at place 89? Chateau? They have an occupied units of 100% but they only have 4 people. So let's drop them

# clearedData = clearedData.drop(index=89, axis=0)

# now let's see what the data looks like

clearedData
```

```python
# The final step is to assign a ranking to all of our neighborhoods.

clearedData.assign(Housing_Rank = 1)
clearedData['Housing_Rank'] = clearedData['% Occupied Units (2010)'].rank(ascending = False)

# let's see what that looks like

clearedData
```

```python
# as we can see, Swisshelm Park is rank 1.0, Ridgemont, 2.0, etc. This will be used later for our final metric.

# We want to visualize this data, but in order to do that, we have to chnage our percentages to a float.

def remove_percentage(x):
    return x.strip('%')

clearedData['% Occupied Units (2010)'] = clearedData['% Occupied Units (2010)'].apply(remove_percentage).astype('float')

# this strips the string of the percent sign and makes it a float for us to work with.

# now let's see what our finalists look like. 

topFifteen = clearedData.head(15)

topFifteen
```

```python
# finally, let's look at what this data looks like visually

units = topFifteen.set_index("Neighborhood")
units.plot(kind='bar', y= '% Occupied Units (2010)', ylim = [93, 97])

units.describe()
```

```python
# Our first place is winner is Swisshelm Park! But, we have to see how it holds up against our other metrics. 
```

```python
# writing list to a file for easier moving around
with open("Conor-ranking.csv", "w+") as f:
    f.write("RANK,NEIGHBORHOOD\n")
    for index, value in clearedData["Neighborhood"].items():
        f.write(f"{index+1},{value}\n")
```
